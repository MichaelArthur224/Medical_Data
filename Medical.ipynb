```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv (r'C:\Users\Michael\Downloads\medical.csv')
df.head()
```

# Plot a histogram with a kde (kernel density estimate)

```python
#The features to analyze: VitD_levels, Doc_visits, TotalCharge
```

# VitD_levels

```python
sns.kdeplot(data = df,  x='VitD_levels');
```

```python
#Add a line for the mean (red)
mean = df['VitD_levels'].mean()
#Add a line for the median (green)
med = df['VitD_levels'].median()
std = df['VitD_levels'].std()
#Add a line for for +1 std from the mean (black)
plus_one_std = mean + std
#Add a line for the - 1 std from the mean (black)
minus_one_std = mean - std
#Highlight the range between +1 and =1 std (yellow)
```

```python
fig,ax = plt.subplots(figsize=(10,6))
sns.histplot(df['VitD_levels'] ,ax=ax, stat='probability')
ax.axvline(mean, color = 'red', label='Mean')
ax.axvline(med, color = 'green', label='Median')
ax.axvline(plus_one_std, color = 'black', label='plus_one_std')
ax.axvline(minus_one_std, color = 'black', ls='--', label='minus_one_std')
ax.axvspan(plus_one_std, minus_one_std, color = 'yellow', zorder = 0)
ax.set_title('VitD_levels')
ax.legend();
```
